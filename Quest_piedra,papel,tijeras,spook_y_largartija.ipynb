```python
import random
import ipywidgets as widgets
from IPython.display import display, HTML

# Opciones del juego con emojis
opciones = {
    "Piedra": "\U0000270A",
    "Papel": "\U0001F44F",
    "Tijera": "\U0000270C",
    "Spock": "\U0001F995",
    "Lagartija": "\U0001F98E"
}

# Reglas del juego
reglas = {
    "Piedra": ["Tijera", "Lagartija"],
    "Papel": ["Piedra", "Spock"],
    "Tijera": ["Papel", "Lagartija"],
    "Spock": ["Tijera", "Piedra"],
    "Lagartija": ["Papel", "Spock"]
}

def jugar(eleccion_usuario):
    eleccion_maquina = random.choice(list(opciones.keys()))

    resultado = ""
    if eleccion_usuario == eleccion_maquina:
        resultado = "\U0001F914 ¬°Empate!"
    elif eleccion_maquina in reglas[eleccion_usuario]:
        resultado = "\U0001F389 ¬°Ganaste!"
    else:
        resultado = "\U0001F62D ¬°Perdiste!"

    mensaje = f"""
    <h2 style='text-align: center;'>
    T√∫ elegiste: {opciones[eleccion_usuario]} ({eleccion_usuario})<br>
    La m√°quina eligi√≥: {opciones[eleccion_maquina]} ({eleccion_maquina})<br>
    <strong>{resultado}</strong>
    </h2>
    """
    display(HTML(mensaje))

# Botones de elecci√≥n
botones = []
for opcion in opciones:
    boton = widgets.Button(description=f"{opciones[opcion]} {opcion}", layout=widgets.Layout(width='150px'))
    boton.style.button_color = "lightblue"
    boton.on_click(lambda b, op=opcion: jugar(op))
    botones.append(boton)

display(widgets.HBox(botones))
```
